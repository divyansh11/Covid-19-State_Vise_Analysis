```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(8,10))
active =sns.barplot(x='Active',y='State/UTs',data=data_state)
active.bar_label(active.containers[0],padding=10,fmt='%g');
plt.show()
```

From the above figure we can see that Kerela has the mostb number of active cases,Moreover the other states including Mah., Tamil nadu also have a significant number of cases.

```python
plt.figure(figsize=(8,10))
total_cases =sns.barplot(x='Total Cases',y='State/UTs',data=data_state)
total_cases.bar_label(total_cases.containers[0],padding=10,fmt='%g',labels=data_state['Total Cases']);
plt.show()
```

From the above graph we can say that Mah. has reported largest number of cases.

```python
plt.figure(figsize=(8,10))
death_ratio=sns.barplot(x='Death Ratio (%)',y='State/UTs',data=data_state)
death_ratio.bar_label(death_ratio.containers[0],padding=10,fmt='%g',label=data_state['Death Ratio (%)'])
plt.show()
```

From the above graph we can say that the number of deaths with total_cases reported happen more in punjab, means covid-19 management of Punjab need to be improve. Also the no. of case reported in uttrakhand is less but the death ratio is more, means the conditions of hospitality and door-to-door monitoring policy need to be implied in the state. 

Moreover, the state which have a death ratio more than 1 need to focus more on their management such that we together can fight in more effective way.
